```python
# 6. Monthly Income Distribution
plt.figure(figsize=(10,5))
plt.hist(df['MonthlyIncome'], bins=25)
plt.title("Distribution of Monthly Income", fontsize=16)
plt.xlabel("Monthly Income")
plt.ylabel("Count")
plt.grid(alpha=0.3)
plt.show()
```

```python
# 7. Attrition Rate by Education Field
plt.figure(figsize=(12,6))
df.groupby('EducationField')['Attrition'].value_counts().unstack().plot(
    kind='bar', figsize=(12,6)
)
plt.title("Attrition Count by Education Field", fontsize=16)
plt.xlabel("Education Field")
plt.ylabel("Count")
plt.legend(title="Attrition")
plt.grid(axis='y', alpha=0.3)
plt.xticks(rotation=45)
plt.show()
```

```python
# 8. Attrition Rate by Marital Status
plt.figure(figsize=(10,5))
df.groupby('MaritalStatus')['Attrition'].value_counts().unstack().plot(
    kind='bar', figsize=(10,5)
)
plt.title("Attrition by Marital Status", fontsize=16)
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# 9. Attrition by Overtime
plt.figure(figsize=(10,5))
df.groupby('OverTime')['Attrition'].value_counts().unstack().plot(
    kind='bar', figsize=(10,5)
)
plt.title("Attrition by OverTime Status", fontsize=16)
plt.xlabel("OverTime (Yes/No)")
plt.ylabel("Count")
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# 10. Average Monthly Income by Job Role
plt.figure(figsize=(12,6))
df.groupby('JobRole')['MonthlyIncome'].mean().sort_values().plot(
    kind='bar', figsize=(12,6)
)
plt.title("Average Monthly Income by Job Role", fontsize=16)
plt.xlabel("Job Role")
plt.ylabel("Average Monthly Income")
plt.grid(axis='y', alpha=0.3)
plt.xticks(rotation=45)
plt.show()
```

```python
# 11. Years at Company Distribution
plt.figure(figsize=(10,5))
plt.hist(df['YearsAtCompany'], bins=15)
plt.title("Distribution of Years at Company", fontsize=16)
plt.xlabel("Years at Company")
plt.ylabel("Frequency")
plt.grid(alpha=0.3)
plt.show()
```

```python
# 12. Job Satisfaction vs Attrition
plt.figure(figsize=(10,5))
df.groupby('JobSatisfaction')['Attrition'].value_counts().unstack().plot(kind='bar')
plt.title("Attrition by Job Satisfaction", fontsize=16)
plt.xlabel("Job Satisfaction Level")
plt.ylabel("Count")
plt.grid(axis='y', alpha=0.3)
plt.xticks(rotation=0)
plt.show()
```

🔹 Job Satisfaction

Low satisfaction (1 or 2) has higher attrition.

Employees with satisfaction = 4 rarely leave.

```python
# 13. Work-Life Balance vs Attrition
plt.figure(figsize=(10,5))
df.groupby('WorkLifeBalance')['Attrition'].value_counts().unstack().plot(kind='bar')
plt.title("Attrition by Work-Life Balance", fontsize=16)
plt.xlabel("Work-Life Balance Rating")
plt.ylabel("Count")
plt.grid(axis='y', alpha=0.3)
plt.xticks(rotation=0)
plt.show()
```

🔹 Work–Life Balance

Poor work-life balance relates strongly to attrition.

Higher ratings show lower attrition.

```python
# 14. Years Since Last Promotion vs Attrition
plt.figure(figsize=(10,5))

# Split data into attrition groups
yes = df[df['Attrition'] == 'Yes']['YearsSinceLastPromotion']
no = df[df['Attrition'] == 'No']['YearsSinceLastPromotion']

plt.hist([yes, no], bins=10, label=['Attrition: Yes', 'Attrition: No'])
plt.title("Years Since Last Promotion vs Attrition", fontsize=16)
plt.xlabel("Years Since Last Promotion")
plt.ylabel("Frequency")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

🔹 Promotions

Employees with no promotion for many years tend to leave more.

# Conclusion

⭐ **HR Policy Recommendations to Improve Employee Retention**

1. **Strengthen Career Growth & Promotion Opportunities**

Employees with no promotion for several years have a higher attrition probability.

Recommended Policies

Implement a transparent promotion cycle (e.g., every 12–18 months).

Create Individual Development Plans (IDPs) for every employee.

Offer career path roadmaps tailored to each Job Role.

Introduce internal job rotation programs (builds engagement + skills).

Set mentorship programs to prepare employees for advancement.

⭐ **2. Improve Job Satisfaction Programs**

Low Job Satisfaction strongly correlates with attrition.

Recommended Policies

Conduct quarterly pulse surveys instead of annual reviews.

Create manager training to improve communication & team leadership.

Implement a reward & recognition platform:

Spot awards

Best performer of the month

Peer recognition badges

Offer project ownership and opportunities for innovation.

⭐** 3. Enhance Work–Life Balance & Reduce Overwork**

Your analysis indicates:

Poor Work-Life Balance → higher attrition

Employees with overtime frequency are more likely to leave

Recommended Policies

Introduce strict work-hour guidelines (avoid consistent overtime).

Provide flexible schedules & remote work options.

Encourage mandatory time-off policies.

Create “No Meeting Fridays” to reduce burnout.

Implement wellness programs (yoga, mental health sessions, counseling).

⭐ **4. Compensation Alignment & Pay Equity**

Certain job roles (Sales, Lab Technicians, HR) showed higher attrition.

Recommended Policies

Perform role-specific salary benchmarking to stay competitive.

Introduce performance-based bonuses.

Offer retention bonuses for critical roles with high turnover.

Conduct pay transparency sessions.

⭐ **5. Improve Manager–Employee Relationship Quality**

People often leave managers, not jobs.

Recommended Policies

Implement manager effectiveness KPIs.

Train managers to:

Reduce micromanagement

Improve feedback quality

Facilitate team collaboration

Introduce skip-level meetings for employees to voice concerns directly to upper management.

⭐ **6. Enhance Onboarding & Early Employee Experience**

Many companies lose employees in their first year.

Recommended Policies

Strengthen the first 90 days with:

Structured onboarding roadmap

Buddy programs

Frequent check-ins

Assign new hire ambassadors in each team.

⭐ 7. Create a Positive Work Culture

Engagement drives retention.

Recommended Policies

Increase team-building activities.

Celebrate small team wins publicly.

Create employee-led committees:

Culture committee

Innovation committee

Diversity & Inclusion committee
